```python
train_data = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
train_data.head(20)
```

```python
test_data = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
test_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
tdb = train_data
```

```python
tdb['Age'] = train_data['Age'].map(lambda x: 0 if x<46.0 else 1 )
```

```python
tdb.head()
```

```python
tedb = test_data
```

```python
tedb['Age'] = test_data['Age'].map(lambda x: 0 if x<46.0 else 1 )
```

```python
tdb.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Transported"]

features = ["HomePlanet", "CryoSleep", "Destination","Age"]
X = pd.get_dummies(tdb[features])
X_test = pd.get_dummies(tedb[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Transported': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
